## pandas 데이터 프레임을 자료구조로 사용
- 파이썬의 딕셔너리를 쉽게 데이터 프레임으로 만들 수 있음.

````python
data ={"name":["nick","judy","alex"]
       ,"age":[10,15,20]};
data
df =pd.DataFrame(data)
df
````

## 기존 열을 활용하여 새로운 열을 쉽게 만들 수 있음.

````python
df['age_plus'] = df['age']+1
df['age_sqared']= df['age']*df['age']
df
````

## join 기능

````python
df2 = pd.DataFrame({ "name": ["nick", "judy", "alex"]
                     ,"height" : [180, 165, 175]
                     , "gender" : ["M", "F", "M"]})
joined = df.set_index("name").join(df2.set_index("name"))
joined
````

## group by 가능

````python
df_mean = joined.groupby("gender").mean()
df_mean
````

4. iloc 사용법
- 행과 열의 위치(정수인덱스)를 사용하여 데이터에 접근
-``` df.iloc[0]``` 첫번째 행 가죠오기
-``` df.iloc[:,0]``` 첫번째 열 가져오기
5.  loc  사용법
- 행과 열의 라벨을 사용하여
-```df.loc[0,'column_name']```
-```df.loc[0:4. colmun_name :'another_column']```

````python
load_krw.iloc[0]
````

````python
load_krw.loc[:,2]
````

````python
load_krw.loc[0,'Name']
````

````python
load_krw.loc[0:4,'Code','Name']
````
